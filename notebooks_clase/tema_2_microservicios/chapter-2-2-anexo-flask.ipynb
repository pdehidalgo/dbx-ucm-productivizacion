# Model Server con Flask

Flask es otro de los frameworks utilizados para exponer microservicios en Python. Ya hemos visto algunas de las bondades de FastAPI, sin embargo, Flask está muy extendido también y es útil que conozcáis ambos y sus diferencias. 

````python
import warnings
warnings.filterwarnings("ignore")

import seaborn as sns
import pandas as pd

from flask import Flask
from flask import jsonify
from flask import request
````

````python
iris = sns.load_dataset("iris")
iris.sample(5, random_state=99)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = iris['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=99)

model = DecisionTreeClassifier(random_state=99)
model.fit(X_train, y_train)
model.score(X_test, y_test)
````

````python
app = Flask(__name__)

@app.route('/prediction', methods=['GET'])
def query():
    data = request.get_json(force=True)
    
    sepal_length = data['sepal_length'],
    sepal_width = data['sepal_width'],
    petal_length = data['petal_length'],
    petal_width = data['petal_width'],
    
    X = pd.DataFrame(zip(sepal_length, sepal_width, petal_length, petal_width))
    predictions = model.predict(X)
    
    pd_result = pd.DataFrame([predictions], columns=["result"])
    
    return jsonify(pd_result.to_json(orient='records'))
````

````python
# Descomentar para levantar el servicio de Flask
# Para acabar el servicio, bastará con interrumpir el kernel o la ejecución de estos comandos

if __name__ == '__main__':
    app.run(port=8080)
````
