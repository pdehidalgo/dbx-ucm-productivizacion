## Creación de un entorno Python con uv

Vamos a crear un entorno Python de cero donde instalar los paquetes que usemos. Para la asignatura, trabajaremos con `uv` como herramienta estándar:

```bash
# 1) Instalar uv
# macOS / Linux:
curl -LsSf https://astral.sh/uv/install.sh | sh

# Windows (PowerShell):
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"

# 2) Comprobar instalación
uv --version

# 3) Crear y activar entorno
uv venv .venv
source .venv/bin/activate   # Git Bash / Linux / macOS
# En PowerShell: .\.venv\Scripts\Activate.ps1

# 4) Instalar dependencias del proyecto
uv add openai python-dotenv ipykernel requests

# 5) Ejecutar comandos dentro del entorno
uv run python --version
uv run python -m ipykernel install --user --name curso-uv
```

> NOTA: si hay algún error, vuelve a lanzar el comando. Si aún así los errores persisten, contacta con el equipo docente.


**uv** es una herramienta para gestionar entornos y dependencias en Python que reemplaza a `pip`, `virtualenv` y `poetry`, con un enfoque en **velocidad, reproducibilidad y simplicidad**: permite crear entornos virtuales, instalar paquetes y resolver dependencias de forma determinista usando un `pyproject.toml` y un lockfile, todo con un rendimiento muy superior (escrito en Rust), lo que lo hace especialmente útil en workflows de CI/CD y MLOps donde necesitas reconstruir entornos de forma rápida y consistente.
