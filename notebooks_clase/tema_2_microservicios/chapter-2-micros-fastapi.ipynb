# Microservicios

Los microservicios son un tipo de **arquitectura de software** en la que una aplicación se construye como una colección de servicios pequeños e independientes que se comunican entre sí mediante APIs bien definidas. Cada microservicio se centra en una funcionalidad específica y puede ser desarrollado, desplegado y escalado de manera independiente. Esta arquitectura contrasta con las aplicaciones monolíticas tradicionales, donde todos los componentes están interconectados y deben ser desplegados como una sola unidad.

`````{admonition} Microservicios
:class: tip
Este enfoque es el estándar en entornos industriales serios, más complicado de orquestar que un proyecto monolítico pero mantenibles y escalables en el largo plazo. 
`````

En otras palabras, los microservicios son un estilo arquitectónico de desarrollo de software que estructura una aplicación como una colección de servicios pequeños, autónomos y modulares. Cada uno de estos servicios ejecuta un proceso único y maneja una parte específica de las funcionalidades del negocio.

## ¿Qué es una API?

Una **API** (Application Programming Interface, o Interfaz de Programación de Aplicaciones) es un conjunto de reglas y definiciones que permiten que dos sistemas de software se comuniquen entre sí.

En el contexto de los **microservicios**, una API es un contrato: define **qué operaciones** ofrece un servicio, **cómo se pueden invocar** (por ejemplo, mediante HTTP/HTTPS), **qué parámetros** se deben enviar y **qué respuestas** se pueden esperar.


## Analogía

Imagina un restaurante como una aplicación software que ha dividido sus operaciones en varios microservicios o pequeños departamentos especializados. Tienes la cocina, el bar, la caja y el servicio de atención al público. Cada uno de estos departamentos hace una tarea específica y esencial:

- La cocina prepara la comida.
- El bar prepara bebidas.
- La caja maneja los pagos.
- El servicio al cliente toma los pedidos y sirve la comida.

Cada uno de estos departamentos funciona de manera independiente pero contribuye al funcionamiento global del restaurante.

Imagina también que la **API** es el **menú** que se ofrece a los clientes. El menú permite a los clientes entender qué se puede pedir (qué operaciones se pueden realizar) y cómo hacer el pedido (cómo interactuar con el servicio).
Por ejemplo, si quieres interactuar con el bar para pedir una bebida, consultas el menú, eliges una bebida y realizas el pedido usando los términos que el menú define. La API del bar define cómo solicitar bebidas, qué parámetros incluir (tipo de bebida, tamaño, etc.), y cómo recibirás tu pedido.

## Los dos tipos más comunes de APIs

- **REST**: basadas en HTTP(*Protocolo de Transferencia de Hipertexto* en internet), muy comunes en microservicios, comunicación unidireccional, un cliente solicita y un servidor entrega una respuesta. 
- **WebSocket**: permite una **comunicación bidireccional en tiempo real** entre un cliente (por ejemplo, un navegador web) y un servidor.

## Ventajas del uso de Microservicios

- **Independencia**: cada microservicio es desarrollado, desplegado, operado y escalado de manera independiente. Esto permite que diferentes equipos trabajen en diferentes servicios con poca o ninguna dependencia entre ellos.
- **Especialización**: cada servicio está diseñado para realizar una función específica. Por ejemplo, un microservicio puede manejar autenticaciones de usuario mientras que otro gestiona las transacciones financieras.
- **Tecnología heterogénea**: los microservicios permiten el uso de diferentes tecnologías y lenguajes de programación dentro de la misma aplicación, según lo que mejor se adapte a cada servicio. Puede que los microservicios del core bancario estén desarrollados en Java pero varios modelos analíticos estén desarrollados en Python o C++.  
- **Resiliencia**: el fallo en un microservicio específico generalmente no afecta el funcionamiento de otros servicios de la aplicación, lo cual mejora la estabilidad del sistema en general.
- **Escalabilidad**: los microservicios se pueden escalar de manera independiente, lo que permite optimizar recursos y mejorar el rendimiento según la demanda específica de cada servicio. Pensemos en microservicios que necesiten ser muy rápidos o que tengan necesidades especiales como uso de GPU. 
- **Despliegue continuo**: al ser independientes, los microservicios facilitan el despliegue continuo y las actualizaciones de la aplicación sin interrumpir el servicio completo.
- **Atomicidad en la ejecución**: también conocido como saga pattern, que es un patrón de diseño para gestionar transacciones distribuidas mediante una secuencia de operaciones independientes coordinadas con eventos o compensaciones en caso de fallo.

## Desafíos en el uso e implementación de Microservicios

- **Complejidad operativa**: requieren herramientas de orquestación, monitoreo y despliegue más sofisticadas.
- **Gestión de datos distribuida**: cada microservicio gestiona la forma en la que consume y recibe datos, lo cual complica las transacciones y la consistencia.
- **Latencia en comunicaciones**: el uso de redes para comunicación entre servicios añade retardo y para determinadas aplicaciones, esto puede suponer un reto técnico.
- **Testing**: las pruebas integradas y de extremo a extremo son más complejas.
- **Versionado y compatibilidad de APIs**: se debe cuidar que los cambios en APIs no rompan la interacción entre servicios. La totalidad de los microservicios de una aplicación no está desarrollada por el mismo equipo por lo que la notificación de cambios y pruebas antes de los despliegues se vuelve imprescindible. Por ejemplo, la implementación de **tensorflow** cambió totalmente de la versión 1.X a 2.X por lo que si el código estaba desarrollado sobre 1.X y subíamos de versión, con total seguridad nuestro microservicio modelo dejaría de funcionar. Hablaremos con más detalle sobre este hecho en el anexo donde tratamos lo que es un monorepo.


## Comunicaciones en API REST: métodos

Los métodos GET, POST, PUT y DELETE, comunes en el protocolo HTTP, son utilizados para comunicarse con servidores y especificar la acción que se desea realizar sobre los recursos del servidor:

- **GET**: utilizado para solicitar datos de un recurso específico. No debe modificar ningún dato en el servidor.
- **POST**: usado para enviar datos al servidor para crear un nuevo recurso. POST también se puede usar para disparar operaciones que no encajan en los otros métodos estándar.
- **PUT**: utilizado para enviar datos a un recurso específico para actualizarlo. Si el recurso no existe, algunos servidores permiten que el método PUT lo cree.
- **DELETE**: usado para eliminar un recurso especificado.

Puede que estos términos os sean familiares bajo el acrónimo de **CRUD** (Create, Read, Update, Delete) que son el conjunto de operaciones básicas para gestionar datos en un sistema informático: crear, leer, actualizar y eliminar.

- **POST /cancel**: cancela un pedido específico. Es utilizado para cancelar un pedido que ya se ha realizado.

## FastAPI

En esta sección aprenderemos a realizar un microservicio en Python que nos permita empaquetar los modelos y que puedan ser consumidos desde el exterior mediante APIs. Uno de los frameworks más populares para crear microservicios en Python es FastAPI, principalmente debido a su simplicidad y flexibilidad.

**FastAPI** es un **framework web moderno y rápido** para construir APIs con Python basado en **tipado estático** (type hints) y **ASGI** (Asynchronous Server Gateway Interface), en concreto, **Starlette**.

Está diseñado para desarrollar **APIs RESTful y WebSocket** de forma eficiente, con soporte nativo para programación asíncrona (`async/await`), validación automática de datos con **Pydantic** y documentación interactiva.

`````{admonition} Más librerías en python
:class: tip
Existen más librerías en python destinadas a crear microservicios como Flask, Tornado, Django, Pyramid o Web2py. Sin embargo, por simplicidad, versatilidad y opciones, FastAPI se está consagrando como un estándar.
`````

`````{admonition} Características clave de FastAPI
:class: info
- **Muy rápido**: su rendimiento se acerca al de Node.js y Go.
- **Validación automática de datos**: gracias a **Pydantic**.
- **Documentación interactiva**: Swagger UI y ReDoc generadas automáticamente.
- **Soporte asíncrono**: ideal para mejorar el rendimiento y uso de WebSocket.
- **Basado en estándares**: OpenAPI (Swagger) y JSON Schema.
- **Ideal para microservicios**: ligero, modular y fácil de escalar.
`````

`````{admonition} Endpoints
:class: important
FastAPI facilita la asociación de URLs con funciones de Python, lo que significa que puedes fácilmente manejar diferentes rutas y endpoints en tu aplicación.
`````



### Ejemplo práctico

A continuación vamos a simular un ejemplo práctico con un modelo desplegado como microservicio. Esta arquitectura es que nos va a permitir diferenciar dos roles de usuarios en los modelos ML: 

- Por un lado estarán los desarrolladores de los modelos ML. Es decir, los encargados del diseño de un modelo dentro del ciclo de vida de ML que una vez consagrado, lo van a marcar para su despliegue y consumo. Para ello, contaremos con una máquina (servidor) que disponibilizará el modelo via API-Rest. 

- Y por otro lado los consumidores de los modelos de ML. Es decir, se preparan unos datos necesarios para poder hacer consumo del modelo desde otra máquina distinta (cliente). Las comunicaciones de ambos roles se harán con sus respectivas máquinas mediante los métodos descritos en la sección de comunicaciones.

Así pues, en una máquina el Data scientist podría diseñar su modelo, entrenarlo y medirlo. Veamos un ejemplo, supongamos que queremos hacer un modelo con el dataset de `iris` que se encuentra en la librería de `seaborn`.

``````python
import warnings
warnings.filterwarnings("ignore")

import seaborn as sns
import pandas as pd

iris = sns.load_dataset("iris")
iris.sample(5, random_state=99)
``````

Supongamos que el modelo final fuera un _decision tree_. Y el modelo lo que pretende es usar las variables `'sepal_length', 'sepal_width', 'petal_length', 'petal_width'` como features predictivas. Así, tenemos:

``````python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = iris['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=99)

model = DecisionTreeClassifier(random_state=99)
model.fit(X_train, y_train)
model.score(X_test, y_test)
``````

Viendo su métrica de error, lo cierto es que nos cuadra y decidimos que éste es el modelo elegido para ponerlo en producción. Esto significaría que un consumidor del modelo podría necesitar una predicción y, para ello, tendría que proporcionar los valores de las variables `'sepal_length', 'sepal_width', 'petal_length', 'petal_width'` y esperar al resultado. 

Por ejemplo, pongamos que los valores sean 6.9, 3.1, 5.4 y 2.1, respectivamente. Para ver la predicción del modelo solo tendríamos que ejecutar la siguiente línea:

``````python
model.predict(pd.DataFrame([[6.9, 3.1, 5.4, 2.1]]))
``````

Sin embargo, de este modo, para poder consumir el modelo como cliente, éste debe estar disponibilizado en nuestro entorno, algo que sencillamente no puede ocurrir por multitud de motivos anteriormente comentados (seguridad, performance, etc). Así pues, **utilizamos FastAPI para poder levantar un servicio en la máquina del servidor donde el modelo esté instanciado pudiendo ejecutar así el `.predict()`**. 

Para poder levantar cualquier servicio de FastAPI, basta con levantar el servicio/aplicación:

```python
from fastapi import FastAPI

app = FastAPI(__name__)

```

## Decorators

Un decorator es una función wrapper que permite modificar el comportamiento de otra función o método, facilitando la extensión y modificación del comportamiento de una función o método. 

- **@app.post("/predict")** es un decorador que permite redireccionar al endpoint predict los métodos POST. Una de las ventajas de FastAPI respecto a Flask (tendréis este mismo ejemplo desarrollado en Flask para que podáis comparar) es su facilidad al usar los decoradores, en Flask, el decorador tendría este aspecto **@app.route("/predict", methods=['POST'])**. 

### ¿Deberíamos usar el método POST o GET?

Dado que vamos a generar una predicción, es una pregunta que podría surgir... Para empezar, estamos enviando datos en el cuerpo de la solicitud (como un JSON con las features incluidas), no en la URL como si sucedería con GET:

```bash
/predict?sepal_length=6.9&sepal_width=3.1&petal_length=5.4&petal_width=2.1
```

POST se utiliza cuando queremos procesar datos que envía el cliente, como en este caso donde hacemos una inferencia. Además, GET no debería tener efectos colaterales ni usarse para operaciones que requieren lógica de procesamiento compleja o que no son **idempotentes** (propiedad de algunas operaciones que garantiza que, sin importar cuántas veces se ejecuten con los mismos datos, el resultado será siempre el mismo). Los modelos de aprendizaje automático son en ocasiones no deterministas por lo que no son siempre idempotentes. 

Por último, FastAPI no permite recibir cuerpos JSON en peticiones GET. Repasando la analogía anterior, podríamos crear decorators para las comunicaciones descritas:

![](../img/microservicios1_handwritten_.png)

```python
from fastapi import FastAPI, Path
from typing import List, Optional

app = FastAPI()

# ---------------------------
# /orders
# ---------------------------

@app.get("/orders")
def get_orders():
    # Lógica para devolver una lista de pedidos
    return {"message": "List of orders"}

@app.post("/orders")
def create_order():
    # Lógica para crear un nuevo pedido
    return {"message": "Order created"}

# ---------------------------
# /orders/{order_id}
# ---------------------------

@app.get("/orders/{order_id}")
def get_order(order_id: int = Path(...)):
    # Lógica para obtener un pedido específico
    return {"message": f"Details of order {order_id}"}

@app.put("/orders/{order_id}")
def update_order(order_id: int):
    # Lógica para actualizar un pedido
    return {"message": f"Order {order_id} updated"}

@app.delete("/orders/{order_id}")
def delete_order(order_id: int):
    # Lógica para eliminar un pedido
    return {"message": f"Order {order_id} deleted"}

# ---------------------------
# /orders/{order_id}/cancel
# ---------------------------

@app.post("/orders/{order_id}/cancel")
def cancel_order(order_id: int):
    # Lógica para cancelar un pedido
    return {"message": f"Order {order_id} cancelled"}

# ---------------------------
# /orders/{order_id}/pay
# ---------------------------

@app.post("/orders/{order_id}/pay")
def pay_order(order_id: int):
    # Lógica para pagar un pedido
    return {"message": f"Order {order_id} paid"}
```

Una vez el servicio FastAPI esté disponbilizado, significa que la máquina donde se levantó el servicio (en nuestro caso, 127.0.0.1 o _localhost_) está preparado para recibir comunicaciones en las siguientes URLs:

- https://127.0.0.1:8000/orders, para devolver la lista de pedidos (GET) o para realizar un pedido nuevo (POST)
- https://127.0.0.1:8000/orders/{order_id}, para devolver la información de un pedido con código _order_id_ (GET), para actualizar un pedido con código _order_id_ (PUT) o para borrar un pedido con código _order_id_ (DELETE)
- https://127.0.0.1:8000/orders/{order_id}/cancel, para cancelar un pedido con código _order_id_ (POST)
- https://127.0.0.1:8000/orders/{order_id}/pay, para pagar un pedido con código _order_id_ (POST)

Volviendo a nuestro ejemplo, podríamos crear una función que realice predicciones del modelo cuando reciba los datos en formato json. 


`````{admonition} Json
:class: info
Los ficheros json son, generalmente, el protocolo clásico para envío de información en todas las comunicaciones de microservicios.
`````


``````python
data = {
    'sepal_length': 6.9,
    'sepal_width': 3.1,
    'petal_length': 5.4,
    'petal_width': 2.1,
}

# Mala praxis: estamos incluyendo el modelo asumiendo que ya existe como variable cuando realmente está en memoria. 
def model_prediction(data):
    
    sepal_length = data['sepal_length'],
    sepal_width = data['sepal_width'],
    petal_length = data['petal_length'],
    petal_width = data['petal_width'],
    
    X = pd.DataFrame(zip(sepal_length, sepal_width, petal_length, petal_width))
    predictions = model.predict(X)
    
    return predictions

model_prediction(data)
``````

Si esta función la wrappeamos con un decorator `/prediction` podríamos estar escuchando a través de una URL los métodos POST (en formato json). Antes de levantar el servicio vamos a explicar en qué consisten las validaciones de datos con FastAPI y Pydantic: 


## Data validation

### ¿Qué es Pydantic?

**Pydantic** es una librería de Python que permite definir modelos de datos usando clases con **tipado estático** (`typing`) y realiza **validación automática** en tiempo de ejecución.

En FastAPI, se usa para definir la **estructura esperada** de las entradas (por ejemplo, el cuerpo de una petición `POST`) y **validar automáticamente** que los datos enviados por el cliente cumplen con el formato correcto. Ejemplo de formato para los datos del ejemplo: 


```python
from pydantic import BaseModel

class IrisFeatures(BaseModel):
    sepal_length: float
    sepal_width: float
    petal_length: float
    petal_width: float

```
>

### Como lo utiliza FastAPI

- Verificar que los tipos de datos son correctos (float, int, str, etc.)

- Garantizar que los campos obligatorios estén presentes. Podríamos añadir campos opcionales si se diera el caso o tipos complejos: 

```python

from pydantic import BaseModel, EmailStr
from typing import Optional

class User(BaseModel):
    name: str
    email: Optional[EmailStr]  # Validación automática del email, campo opcional
```

- Lanza errores claros (HTTP 422) cuando los datos no cumplen los requisitos.

- Convertir automáticamente datos JSON a objetos Python.

- Documentar automáticamente los parámetros esperados en la API para incluirlos en el Swagger.


``````python
# main.py
from fastapi import FastAPI
from pydantic import BaseModel, field_validator
import pandas as pd

# Asumimos que el modelo está en memoria puesto que hemos ejecutado las anteriores celdas primero

app = FastAPI()

# Definir esquema de entrada
class IrisFeatures(BaseModel):
    sepal_length: float
    sepal_width: float
    petal_length: float
    petal_width: float

    # Validación al vuelo del contenido de los datos, podría hacerse para todos los inputs desacoplando esta acción del código de predicción
    @field_validator('sepal_length')
    @classmethod
    def check_sepal_length(cls, v):
        if v <= 0:
            raise ValueError("sepal_length must be positive")
        return v

# Función de predicción
def model_prediction(data: dict):
    X = pd.DataFrame([[
        data['sepal_length'],
        data['sepal_width'],
        data['petal_length'],
        data['petal_width'],
    ]], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    
    predictions = model.predict(X)
    return predictions[0]

# Decorador FastAPI que expone el endpoint
@app.post("/predict")
def predict(features: IrisFeatures):
    data = features.model_dump()
    prediction = model_prediction(data)
    return {"prediction": prediction}
``````

### Model Consumption (Model Client)

Supongamos ahora otra máquina distinta, ubicada incluso en otra región y país. En esta máquina al cliente que quiere consumir el modelo. 

Existen varias opciones. Desde python utilizando la librería requests: 

``````python
import requests

data = {
    'sepal_length': 6.9,
    'sepal_width': 3.1,
    'petal_length': 5.4,
    'petal_width': 2.1,
}

# Usa POST, no GET
response = requests.post("http://127.0.0.1:8000/predict", json=data)

# Muestra la respuesta como JSON
response.json()
``````

Desde python como un subproceso y empleando el comando CURL (también podríais lanzarlo desde un terminal bash con cURL instalado o desde un cliente como **Postman**): 

``````python
import subprocess

# Permite manejar el output
response = subprocess.run(
    [
        "curl", "-X", "POST", "http://127.0.0.1:8000/predict",
        "-H", "Content-Type: application/json",
        "-d", '{"sepal_length": 6.9, "sepal_width": 3.1, "petal_length": 5.4, "petal_width": 2.1}'
    ],
    capture_output=True,
    text=True
)

print("Status:", response.returncode)
print("Respuesta:", response.stdout)
``````

`````{admonition} Orden de los procesos
:class: important
Si el servicio FastAPI no está levantado y, por tanto, escuchando en el lado del Server, no podrá ser consumido en el lado del cliente y fallará.
`````


`````{admonition} Códigos de Error
:class: info
Si el proceso funciona correctamente, el status que devuelve como respuesta el servicio es 200. En otro caso, podría haber fallado por los siguientes motivos principales:
- Response [200] -> todo OK
- Response [400] -> no alcanza al server
- Response [422] -> los datos a la entrada no son correctos
- Response [500] -> algún error en el server
`````

``````python
# Probemos con un sepal length negativo

data = {
    'sepal_length': -6.9,
    'sepal_width': 3.1,
    'petal_length': 5.4,
    'petal_width': 2.1,
}

response = requests.post("http://127.0.0.1:8000/predict", json=data)

# Muestra la respuesta como JSON
response.json()
``````

Observamos que el error es suficientemente informativo como para tomar acción y cambiar el valor desde el cliente. Estas validaciones son extremandamente útiles cuando hablamos de cientos de features a la entrada en un entorno real con cientos o miles de peticiones por minuto.
